```python
tables = pd.read_html(url)
tables
```

```python
tables1 = pd.read_html(url1)
tables1
```

```python
df15 = tables1[0]
df15['Year'] = '2015'
df15.head()
```

```python
df16 = pd.read_csv(file)
df16['Year'] ='2016'
df16.head()
```

```python
df17 = tables[0]
df17["Year"] = '2017'
df17.head()
```

```python
frames = [df15, df16, df17]
df3years= pd.concat(frames)
df3years.tail()
```
